```python
import json
import networkx as nx

def loadgraph1(filename = "graph1.json"):
    # Charger le JSON d'un graphe avec Nodes + Edges séparés
    with open(filename, "r") as f:
        data = json.load(f)

    G = nx.DiGraph()
    for node in data["nodes"]:
        G.add_node(node["id"], duration=node["duration"])
    for edge in data["edges"]:
        G.add_edge(edge["source"], edge["target"])

    print(G.nodes(data=True))
    print(G.edges())

loadgraph1()
```
